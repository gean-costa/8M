```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from wordcloud import WordCloud
```

```python
hashtags_df = pd.read_parquet('../EDA/hashtags.parquet.gzip')
data = dict(zip(hashtags_df['hashtag'].tolist(), hashtags_df['count'].tolist()))

# x, y = np.ogrid[:300, :300]
# mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
# mask = 255 * mask.astype(int)

# Generate a word cloud image
word_cloud = WordCloud(background_color='white', 
                       width=1280, 
                       height=720,
                       max_words=200,
                       relative_scaling=0).generate_from_frequencies(data)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (15,15))
plt.axis("off")
plt.imshow(word_cloud, interpolation = 'bilinear')
plt.show()

word_cloud.to_file('wordcloud-hashtags.png')
```
