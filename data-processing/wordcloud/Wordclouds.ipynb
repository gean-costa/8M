```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## PortuguÃªs

```python
with open('../2-data-pre-processing/cenario1_8M2020_tweets_pt.txt', 'r') as file:
    tweets_pt = file.readlines()
    
palavras = ','.join(list(tweets_pt))

# Generate a word cloud image
nuvem_palavras = WordCloud(background_color='white', width=1280, height=720).generate(palavras)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (15,15))
plt.axis("off")
plt.imshow(nuvem_palavras, interpolation = 'bilinear')
plt.show()

nuvem_palavras.to_file('wordcloud-pt-1.png')
```
